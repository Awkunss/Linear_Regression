```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train = np.array([
    1.1, 1.3, 1.5, 2.0, 2.2, 2.9, 3.0, 3.2, 3.2, 3.7, 
    3.9, 4.0, 4.0, 4.1, 4.5, 4.9, 5.1, 5.3, 5.9, 6.0, 
    6.8, 7.1, 7.9, 8.2, 8.7, 9.0, 9.5, 9.6, 10.3, 10.5
])

y_train=np.array([
    39343.00, 46205.00, 37731.00, 43525.00, 39891.00, 
    56642.00, 60150.00, 54445.00, 64445.00, 57189.00, 
    63218.00, 55794.00, 56957.00, 57081.00, 61111.00, 
    67938.00, 66029.00, 83088.00, 81363.00, 93940.00, 
    91738.00, 98273.00, 101302.00, 113812.00, 109431.00, 
    105582.00, 116969.00, 112635.00, 122391.00, 121872.00
])
```

```python
# Hàm tạo đường thẳng dự đoán từ b1, b0
def calculate_cost(b1, b0, x_train):
    m = x_train.shape[0]
    cost_list = np.zeros(m)
    for i in range(m):
        cost_list[i] = (b1 * x_train[i] + b0)
    return cost_list
```

# Tìm B0, B1 bằng Gradient Descent

```python
m = x_train.shape[0]
```

```python
b1 = 0.1
b0 = 0.1
```

```python
plt.scatter(x_train, y_train, marker="x", c="r")
plt.title("Lương theo số năm kinh nghiệm")
plt.xlabel("Danh sách số năm kinh nghiệm")
plt.ylabel("Lương")
plt.show()
```

```python
def calculate_deriative(b1, b0, tmp_alpha, x_train, y_train):
    m = x_train.shape[0]
    sum_deriative_b1 = 0
    sum_deriative_b0 = 0
    for i in range(m):
        sum_deriative_b1 += (b1 * x_train[i] + b0 - y_train[i]) * x_train[i]
        sum_deriative_b0 += (b1 * x_train[i] + b0 - y_train[i])
    return tmp_alpha * 1 / m * sum_deriative_b1, tmp_alpha * 1 / m * sum_deriative_b0
```

```python
def calculate_b1_and_b0(b1, b0, tmp_alpha ,x_train, y_train):
    num_iters = 10000
    for i in range(num_iters):
        deri_b1, deri_b0 = calculate_deriative(b1, b0, tmp_alpha, x_train, y_train)
        b1 = b1 - deri_b1
        b0 = b0 - deri_b0
    return b1, b0
```

```python
tmp_alpha = 0.01
b1, b0 = calculate_b1_and_b0(b1, b0, tmp_alpha, x_train, y_train)
```

```python
print(b1,b0)
```

```python
cost_list = calculate_cost(b1, b0, x_train)
plt.scatter(x_train, y_train, marker="x", c="r", label="Atual Values")
plt.title("Lương theo số năm kinh nghiệm")
plt.xlabel("Danh sách số năm kinh nghiệm")
plt.ylabel("Lương")
plt.plot(x_train, cost_list, c="b", label="Prediction Line")
plt.show()
```

# Tìm B0, B1 bằng đạo hàm


```python
mean_x = np.mean(x_train)
mean_y = np.mean(y_train)
print(mean_x, mean_y)
for i in range(m):
    b1cal = (x_train[i] - mean_x) * (y_train[i] - mean_y)/(x_train[i] - mean_x)**2
    b0cal = mean_y - b1cal * mean_x
print(b1cal, b0cal)
```

```python
cost_list = calculate_cost(b1cal, b0cal, x_train)
plt.scatter(x_train, y_train, marker="x", c="r", label="Atual Values")
plt.title("Lương theo số năm kinh nghiệm")
plt.xlabel("Danh sách số năm kinh nghiệm")
plt.ylabel("Lương")
plt.plot(x_train, cost_list, c="b", label="Prediction Line")
plt.show()
```
